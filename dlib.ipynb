```python
# Импорт необходимых библиотек
import face_recognition
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files
```

```python
# Функция для распознавания лиц с использованием библиотеки Face Recognition
def recognize_faces(image_path):
    # Загрузка изображения
    image = cv2.imread(image_path)

    # Поиск всех обнаруженных лиц на изображении
    face_locations = face_recognition.face_locations(image)

    # Отображение количества найденных лиц
    print(f"Found {len(face_locations)} face(s) in the image.")

    # Отображение изображения с ограничивающими рамками вокруг обнаруженных лиц
    for face_location in face_locations:
        top, right, bottom, left = face_location
        cv2.rectangle(image, (left, top), (right, bottom), (0, 255, 0), 2)

    # Конвертация изображения из формата BGR в RGB (для отображения с помощью matplotlib)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Отображение изображения с ограничивающими рамками
    plt.figure(figsize=(8, 8))
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.show()

# Загрузка изображения в Colab
uploaded = files.upload()

# Получение пути загруженного изображения
image_path = next(iter(uploaded.keys()))

# Распознавание лиц на загруженном изображении
recognize_faces(image_path)
```
